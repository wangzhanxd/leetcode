### 242. 有效的字母异位词
给定两个字符串 s 和 t ，编写一个函数来判断 t 是否是 s 的字母异位词。

示例 1:

输入: s = "anagram", t = "nagaram"
输出: true
示例 2:

输入: s = "rat", t = "car"
输出: false
说明:
你可以假设字符串只包含小写字母。

进阶:
如果输入字符串包含 unicode 字符怎么办？你能否调整你的解法来应对这种情况？

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        s_l = [0]*26
        t_l = [0]*26
        for ele in s:
            s_l[ord(ele)-ord('a')] = s_l[ord(ele)-ord('a')] + 1
        for ele in t:
            t_l[ord(ele)-ord('a')] = t_l[ord(ele)-ord('a')] + 1
        return s_l == t_l
```

### 409. 最长回文串
给定一个包含大写字母和小写字母的字符串，找到通过这些字母构造成的最长的回文串。

在构造过程中，请注意区分大小写。比如 "Aa" 不能当做一个回文字符串。

注意:
假设字符串的长度不会超过 1010。

示例 1:

输入:
"abccccdd"

输出:
7

解释:
我们可以构造的最长的回文串是"dccaccd", 它的长度是 7。

```python
class Solution:
    def longestPalindrome(self, s: str) -> int:
        cnts = [0]*256
        for ele in s:
            cnts[ord(s)] += 1
        ret = 0
        for ele in cnts:
            ret += int(cnt/2)*2
        return ret if ret==len(s) else ret+1
```

### 205. 同构字符串
给定两个字符串 s 和 t，判断它们是否是同构的。

如果 s 中的字符可以被替换得到 t ，那么这两个字符串是同构的。

所有出现的字符都必须用另一个字符替换，同时保留字符的顺序。两个字符不能映射到同一个字符上，但字符可以映射自己本身。

示例 1:

输入: s = "egg", t = "add"
输出: true
示例 2:

输入: s = "foo", t = "bar"
输出: false
示例 3:

输入: s = "paper", t = "title"
输出: true
    
说明:
你可以假设 s 和 t 具有相同的长度。

```python
# 用list记录前一次字符出现的位置
class Solution:
    def isIsomorphic(self, s: str, t: str) -> bool:
        ss = [0]*256
        st = [0]*256
        for i in range(len(s)):
            if ss[ord(s[i])] != st[ord(t[i])]:
                return False
            st[ord(t[i])] = ss[ord(s[i])] = i+1   ###不能是i，必须是i+一个数，否则ab aa返回是True
        return True
```

### 647. 回文子串
给定一个字符串，你的任务是计算这个字符串中有多少个回文子串。

具有不同开始位置或结束位置的子串，即使是由相同的字符组成，也会被计为是不同的子串。

示例 1:

输入: "abc"
输出: 3
解释: 三个回文子串: "a", "b", "c".
示例 2:

输入: "aaa"
输出: 6
说明: 6个回文子串: "a", "a", "a", "aa", "aa", "aaa".
注意:

输入的字符串长度不会超过1000。

```python
###方法- 暴力解法 时间超时 考虑使用递归求子序列
class Solution:
    def isPalindromic(self,s):
        i = 0
        j = len(s)-1
        while i<=j:
            if s[i]!=s[j]:
                return False
            i += 1
            j -= 1
        return True
    def countSubstrings(self, s: str) -> int:
        cnt = 0
        n = len(s)
        for i in range(n):
            for j in range(n+1):
                if j>i:
                    cnt += self.isPalindromic(s[i:j])
        return cnt
```

```python
### 方法二 从任一位置向外扩展 计算回文串个数
class Solution:
    def __init__(self):
        self.cnt = 0
    def extendSubstrings(self,s,start,end):
        while start>=0 and end<len(s) and s[start]==s[end]:
            start -= 1
            end += 1
            self.cnt += 1
    def countSubstrings(self, s: str) -> int:
        for i in range(len(s)):
            self.extendSubstrings(s,i,i)  ###奇数个子串
            self.extendSubstrings(s,i,i+1)  ###偶数个子串
        return self.cnt
```

### 696. 计数二进制子串
给定一个字符串 s，计算具有相同数量0和1的非空(连续)子字符串的数量，并且这些子字符串中的所有0和所有1都是组合在一起的。

重复出现的子串要计算它们出现的次数。

示例 1 :

输入: "00110011"
输出: 6
解释: 有6个子串具有相同数量的连续1和0：“0011”，“01”，“1100”，“10”，“0011” 和 “01”。

请注意，一些重复出现的子串要计算它们出现的次数。

另外，“00110011”不是有效的子串，因为所有的0（和1）没有组合在一起。
示例 2 :

输入: "10101"
输出: 4
解释: 有4个子串：“10”，“01”，“10”，“01”，它们具有相同数量的连续1和0。
注意：

s.length 在1到50,000之间。
s 只包含“0”或“1”字符。

```python
## 方法一 同647：回文串 任一位置向外扩展
class Solution:
    def __init__(self):
        self.cnt = 0
    def helper(self,s,start,end):
        while start>=0 and end<len(s):
            if s[start]!=s[end] and end-start==1:
                start -= 1
                end += 1
                self.cnt += 1
            elif (end-start)>1 and s[start]==s[start+1] and s[end]==s[end-1]:
                start -= 1
                end += 1
                self.cnt += 1
            else:
                break  
    def countBinarySubstrings(self, s: str) -> int:
        for i in range(len(s)):
            self.helper(s,i,i+1)
        return self.cnt
```

```python
### 方法二 一遍遍历 两个数分别记录当前1的数目和0的数目
### curLen是指当前遍历区域0或1的数目
class Solution:
    def countBinarySubstrings(self, s: str) -> int:
        preLen = 0
        curLen = 1
        cnt = 0
        for i in range(1,len(s)):
            if s[i] == s[i-1]:
                curLen += 1
            else:
                preLen = curLen
                curLen = 1
            if curLen < preLen:
                cnt += 1
        return cnt
```
